```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
areadf = pd.read_csv('data/area.csv')
```

```python
areadf.plot.scatter('area','price')
```

```python
X = areadf[['area']].values
y = areadf['price'].values
```

```python
X,y
```

```python
lr = LinearRegression()
```

```python
model = lr.fit(X,y)
```

```python
print("The Predicted value for Area (1037)",model.predict(np.array([[972]])))
```

```python
m = model.coef_
C = model.intercept_
print("Slope for the Straight Line",m)
print("Intercept for the Straight Line",C)
```
